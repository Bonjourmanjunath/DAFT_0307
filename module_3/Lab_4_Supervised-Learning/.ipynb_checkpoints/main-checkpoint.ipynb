```python
# Your code here


# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()

# Display the initial heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Initial Correlation Heatmap")
plt.show()

# Find the highest correlated pair
max_corr = correlation_matrix.abs().max()
high_corr_cols = max_corr[max_corr > 0.9].index

# Remove one column from each correlated pair
for col1, col2 in zip(high_corr_cols[:-1], high_corr_cols[1:]):
    if correlation_matrix.loc[col1, target_column] > correlation_matrix.loc[col2, target_column]:
        websites.drop(col1, axis=1, inplace=True)
    else:
        websites.drop(col2, axis=1, inplace=True)
    print(f"Removed column: {col1 if correlation_matrix.loc[col1, target_column] > correlation_matrix.loc[col2, target_column] else col2}")

# Calculate the new correlation matrix
correlation_matrix = websites.corr()

# Display the updated heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Updated Correlation Heatmap")
plt.show()
```

# Challenge 4 - Handle `WHOIS_*` Categorical Data

There are several categorical columns we need to handle. These columns are:

* `URL`
* `CHARSET`
* `SERVER`
* `WHOIS_COUNTRY`
* `WHOIS_STATEPRO`
* `WHOIS_REGDATE`
* `WHOIS_UPDATED_DATE`

How to handle string columns is always case by case. Let's start by working on `WHOIS_COUNTRY`. Your steps are:

1. List out the unique values of `WHOIS_COUNTRY`.
1. Consolidate the country values with consistent country codes. For example, the following values refer to the same country and should use consistent country code:
    * `CY` and `Cyprus`
    * `US` and `us`
    * `SE` and `se`
    * `GB`, `United Kingdom`, and `[u'GB'; u'UK']`

#### In the cells below, fix the country values as intructed above.

```python
# Your code here




unique_countries = websites['whois_country'].unique()

# Print the unique country values before consolidation
print("Unique countries before consolidation:")
print(unique_countries)

# Dictionary to map inconsistent values to consistent country codes
country_mapping = {
    'CY': 'Cyprus',
    'us': 'US',
    'Cyprus': 'CY',
    'se': 'SE',
    'United Kingdom': 'GB',
    "[u'GB'; u'UK']": 'GB'
}

# Apply the mapping to consolidate country values
websites['whois_country'] = websites['whois_country'].replace(country_mapping)

# Print the unique country values after consolidation
print("\nUnique countries after consolidation:")
print(websites['whois_country'].unique())
```

Since we have fixed the country values, can we convert this column to ordinal now?

Not yet. If you reflect on the previous labs how we handle categorical columns, you probably remember we ended up dropping a lot of those columns because there are too many unique values. Too many unique values in a column is not desirable in machine learning because it makes prediction inaccurate. But there are workarounds under certain conditions. One of the fixable conditions is:

#### If a limited number of values account for the majority of data, we can retain these top values and re-label all other rare values.

The `WHOIS_COUNTRY` column happens to be this case. You can verify it by print a bar chart of the `value_counts` in the next cell to verify:

```python
# Your code here


import matplotlib.pyplot as plt

# Plotting the value counts of WHOIS_COUNTRY
plt.figure(figsize=(12, 6))
websites['whois_country'].value_counts().plot(kind='bar')
plt.title("Value Counts of whois_country")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

#### After verifying, now let's keep the top 10 values of the column and re-label other columns with `OTHER`.

```python
# Your code here

# Get the top 10 most common countries
top_countries = websites['whois_country'].value_counts().head(10).index

# Replace values not in the top 10 with 'OTHER'
websites.loc[~websites['whois_country'].isin(top_countries), 'whois_country'] = 'OTHER'

# Verify the changes
print("Unique values in whois_country after re-labeling:")
print(websites['whois_country'].unique())
```

Now since `WHOIS_COUNTRY` has been re-labelled, we don't need `WHOIS_STATEPRO` any more because the values of the states or provinces may not be relevant any more. We'll drop this column.

In addition, we will also drop `WHOIS_REGDATE` and `WHOIS_UPDATED_DATE`. These are the registration and update dates of the website domains. Not of our concerns.

```python
# Your code here

unique_server_values = websites['server'].unique()

print(unique_server_values)
```

Although there are so many unique values in the `SERVER` column, there are actually only 3 main server types: `Microsoft`, `Apache`, and `nginx`. Just check if each `SERVER` value contains any of those server types and re-label them. For `SERVER` values that don't contain any of those substrings, label with `Other`.

At the end, your `SERVER` column should only contain 4 unique values: `Microsoft`, `Apache`, `nginx`, and `Other`.

```python
# Your code here


# Define the main server types
main_server_types = ['Microsoft', 'Apache', 'nginx']

# Function to consolidate server values
def consolidate_server(server_value):
    for server_type in main_server_types:
        if server_type.lower() in server_value.lower():
            return server_type
    return 'Other'

# Apply the consolidation function to the SERVER column
websites['server'] = websites['server'].apply(consolidate_server)

# Print the unique values in SERVER column after consolidation
unique_server_values_after = websites['server'].unique()
print("\nUnique values in SERVER column after consolidation:")
print(unique_server_values_after)
```

```python
# Count `SERVER` value counts here



server_value_counts = websites['server'].value_counts()
server_value_counts
```

# Challenge 6 - Modeling, Prediction, and Evaluation

We'll start off this section by splitting the data to train and test. **Name your 4 variables `X_train`, `X_test`, `y_train`, and `y_test`. Select 80% of the data for training and 20% for testing.**

```python
from sklearn.model_selection import train_test_split

# Your code here:



X = website_dummy[['url_length','tcp_conversation_exchange','dist_remote_tcp_port','remote_ips','app_bytes','source_app_packets','remote_app_packets','source_app_bytes','remote_app_bytes','app_packets','whois_country_CA','whois_country_ES','whois_country_GB','whois_country_IN','whois_country_JP','whois_country_None','whois_country_OTHER','whois_country_PA','whois_country_UK','whois_country_US']]
Y = website_dummy['type']
# Split the data into train and test sets (80% train, 20% test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Print the shapes of the train and test sets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", Y_train.shape)
print("Shape of y_test:", Y_test.shape)
```

#### In this lab, we will try two different models and compare our results.

```python
# Your code here:

logreg_model = LogisticRegression()
```

Next, fit the model to our training data. We have already separated our data into 4 parts. Use those in your model.

```python
# Your code here:

logreg_model.fit(X_train, Y_train)
```

finally, import `confusion_matrix` and `accuracy_score` from `sklearn.metrics` and fit our testing data. Assign the fitted data to `y_pred` and print the confusion matrix as well as the accuracy score
